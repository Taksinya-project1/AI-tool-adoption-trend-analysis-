Phase 5-Hypothesis Testing

```python
from scipy.stats import ttest_ind
```

```python
small = df[df['company_size'] == 'Small']['productivity_change_percent']

large = df[df['company_size'] == 'Large']['productivity_change_percent']
```

```python
from scipy.stats import ttest_ind

startup = df[df['company_size'] == 'Startup']['productivity_change_percent']
enterprise = df[df['company_size'] == 'Enterprise']['productivity_change_percent']

t_stat, p_value = ttest_ind(startup, enterprise)

print("T-Statistic =", t_stat)
print("P-Value =", p_value)

if p_value < 0.05:
    print("Reject Null Hypothesis (H0)")
    print("There is a significant difference in productivity between Startup and Enterprise companies.")
else:
    print("Fail to Reject Null Hypothesis (H0)")
    print("There is no significant difference in productivity between Startup and Enterprise companies.")
```

Phase 6-Organization segmentation

```python
def segment_company(adoption_rate):
    if adoption_rate >= 80:
        return "AI Leader"
    elif adoption_rate >= 50:
        return "Early Adopter"
    else:
        return "Slow Adopter"

df["organization_segment"] = df["ai_adoption_rate"].apply(segment_company)

df.head()
```

```python
print(df["organization_segment"].value_counts())
```
